```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../data/input/titanic_data.csv')
data
```

```python
data = data.drop(columns='PassengerId', inplace=False)
data
```

```python
y_train = data['Survived'].values
x_train = data.drop(columns='Survived', inplace=False).values

x_train, y_train
```

```python
x_test = x_train[600:714]
y_test = y_train[600:714]
x_train = x_train[0:600]
y_train = y_train[0:600]

x_test.shape, y_test.shape
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential(
    [
        keras.Input(shape=(6,)),
        layers.Dense(5, activation='sigmoid', name='layer_1'),
        layers.Dense(2, activation='sigmoid', name='layer_2'),
        layers.Dense(1, activation='sigmoid', name='layer_3'),
    ],
    name='MLP_titanic'
)

model.build()
model.summary()
```

```python
from tensorflow.keras import optimizers
from tensorflow.keras import losses
from tensorflow.keras import metrics

opt = optimizers.SGD(learning_rate=0.3)
f_loss = losses.BinaryCrossentropy()
met = ['accuracy']

model.compile(
    optimizer=opt,
    loss=f_loss,
    metrics=met
)
```

```python
model.fit(
    x=x_train,
    y=y_train,
    batch_size=60,
    epochs=300
)
```
